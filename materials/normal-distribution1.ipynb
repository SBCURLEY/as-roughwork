```python
# Flip a single coin several times.
flips = np.random.binomial(1, 0.5, 1000)  # 1000 flips of a fair coin

# Show.
flips
```

```python
# Count heads.  heads is a 1 and tails is a 0
heads = flips.sum()   # takes teh flips array and sums it up    .sum() is in numpy array    

# Show.
heads
```

#### 1000 Coins
Instead of flipping one coin 1000 times, flip 1000 coins once

You are simulating the total number of heads from 1000 coin flips. The value stored in flips is already an integer that represents the number of heads you got in those 1000 flips (for example, 500 heads). There's no need to sum anything because flips is already the sum.
So, in this case, flips is already the result you want — you don't need to sum it again. You can simply use flips directly as the number of heads.

```python
np.random.binomial(1000, 0.5) 
```

```python
flips = np.random.binomial(1000, 0.5)  
# Show.
print(f"In this simulation, out of 1000 flips, {flips} were heads.")
```

Clarification:

To recap:

np.random.binomial(1000, 0.5) gives you a single integer, which is the number of heads (successes) from 1000 trials.

If you wanted to simulate each individual flip (0 or 1 for each coin flip), you’d use np.random.binomial(1, 0.5, 1000).

```python
# Create a blank plot. discrete uniform distribution
fig, ax = plt.subplots()

# Create a bar chart.
ax.bar([0, 1], [1000 - heads, heads])   #1000 minus heads  = taills

# Add x labels.
ax.set_xticks([0, 1], labels=['Tails', 'Heads'])

# Add title.
ax.set_title('Flipping a coin 1000 times')

# Add axis label.
ax.set_xlabel('No heads')
ax.set_ylabel('Count');
```

### Flipping Two Coins

```python
# Flip two coins several times.
# 0 = no heads, 1 = one head, 2 = two heads. Where 1 is getting one head and one tail. We dont know which is which
# 2 experiments of flipping two coins, each with 0.5 probability, do it a thousand times
# because of the above 2, it made sure results are between 0 and 2
# 1 loses information about which coin was heads and which was tails
twoflips = np.random.binomial(2, 0.5, 1000)

# Show.
twoflips
```

```python
# Count the number of times each value occurs. Earlier w used teh sum function.
# add urllity to get unique values and their counts
# ratios look suspicious - why is 1 coming out twice as likely as 2 or 0?
uniques, counts = np.unique(twoflips, return_counts=True)

# Show.
uniques, counts
```

```python
# Create a blank plot.
fig, ax = plt.subplots()

# Create a bar chart.
ax.bar(uniques, counts)

# Set x ticks.
ax.set_xticks(uniques)

# Add title.
ax.set_title('Flipping two coins 1000 times')

# Add axis label.
ax.set_xlabel('No heads')
ax.set_ylabel('Count');
```
