## Probability Density Function

$ f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $

https://en.wikipedia.org/wiki/Normal_distribution

```python
def f(x, mu=0.0, sigma=1.0):
  # The normalisation constant.
  norm = 1.0 / np.sqrt(2.0 * np.pi * sigma**2)
  
  # f(x)
  return norm * np.exp(-(x - mu)**2 / (2.0 * sigma**2))
```

```python
# Test the function.
f(0.0)
```

```python
# Create an x range.
x = np.linspace(-5.0, 5.0, 1001)

# Get y values.
y = f(x)

# Show.
x, y
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Plot the function.
ax.plot(x, y, 'r');
```

Reference: [6.4: Normal Approximation to the Binomial Distribution](https://stats.libretexts.org/Courses/Las_Positas_College/Math_40%3A_Statistics_and_Probability/06%3A_Continuous_Random_Variables_and_the_Normal_Distribution/6.04%3A_Normal_Approximation_to_the_Binomial_Distribution)

```python
# Number of coins to flip.
n = 100

# Number of times to do it.
k = 10000 * n

# Probability of heads.
p = 0.5

# Flip two coins several times.
nflips = np.random.binomial(n, p, k)

# Count the number of times each value occurs.
uniques, counts = np.unique(nflips, return_counts=True)

# Create a blank plot.
fig, ax = plt.subplots(figsize=(12, 5))

# Create a bar chart.
ax.bar(uniques, counts / k)

# Create an x range.
x = np.linspace(25.0, 75.0, 1001)

# Get y values.
y = f(x, mu=n*p, sigma=np.sqrt(n * p * (1 - p)))

# Plot the function.
ax.plot(x, y, 'r-')

# Set x ticks.
ax.set_xticks(uniques)

# Add title.
ax.set_title(f'Flipping {n} coins {k} times')

# Add axis label.
ax.set_xlabel('No heads')
ax.set_ylabel('Proportion');
```

## Sampling Distribution

```python
# Generate a random normal sample.
# rows = 100000  columns = 100
# start out with 10 million values from standard normal distribution (centered on zero sd = 1)
# generated 10 million numbers, put them into sets of 100 to make 100000 rows
# Within each set of 100 numbers, we will calculate the mean 9add all 100 numbers and divided by 100 to get the mean)
data = np.random.normal(size=(100000, 100))

# Calculate mean.
# numpy.mean takes the grid from the specified axis.
# axis=1 means calculate the mean for each row. 100 numbers per row.
# axis=0 would calculate the mean for each column. 100000 numbers per column.
means = np.mean(data, axis=1)

# Show.
means
```

```python
# Create a blank plot.
fig, ax = plt.subplots()

# Create a histogram.
ax.hist(means, edgecolor='black', density=True, bins=np.linspace(-0.4, 0.4, 33));
```

```python
# Create a blank plot.
fig, ax = plt.subplots()

# Create a histogram.
ax.hist(means, edgecolor='black', density=True, bins=np.linspace(-0.4, 0.4, 33))

# Create an x range.
x = np.linspace(-0.4, 0.4, 1001)

# Get y values.
y = f(x, mu=0.0, sigma=1.0 / np.sqrt(100))

# Plot the function.
ax.plot(x, y, 'r-');
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Simulation ---
N_SAMPLES = 100000
SAMPLE_SIZE = 10

# draw 100k samples of size 10
samples = np.random.normal(loc=0, scale=1, size=(N_SAMPLES, SAMPLE_SIZE))

# compute SD with ddof=0 (population) and ddof=1 (sample)
sd_pop = np.std(samples, axis=1, ddof=0)
sd_sample = np.std(samples, axis=1, ddof=1)

# --- Step 2: Plotting ---
fig, ax = plt.subplots(figsize=(8,5))

bins = np.linspace(0, 2, 80)

ax.hist(sd_pop, bins=bins, density=True, alpha=0.6, label='Population SD (ddof=0)')
ax.hist(sd_sample, bins=bins, density=True, alpha=0.6, label='Sample SD (ddof=1)')

ax.set_xlabel("Standard deviation")
ax.set_ylabel("Density")
ax.set_title("Population SD vs Sample SD (n = 10)")
ax.legend()

plt.show()
```
