## Probability Density Function

$ f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $

https://en.wikipedia.org/wiki/Normal_distribution

```python
def f(x, mu=0.0, sigma=1.0):
  # The normalisation constant.
  # first part of equation above  
  norm = 1.0 / np.sqrt(2.0 * np.pi * sigma**2)
  
  # f(x)
  return norm * np.exp(-(x - mu)**2 / (2.0 * sigma**2))
```

```python
# Test the function.
f(0.0)
```

```python
# Like red line on wikipedia page
# when you go up on bell curve, and move across to y - it is around 0.4
```

```python
# Create an empty plot.
# recreation of red curve on wikipedia page
fig, ax = plt.subplots()

# Plot the function.
ax.plot(x, y, 'r');
```

```python
# Number of coins to flip.
n = 100

# Number of times to do it. a million times
k = 10000 * n

# Probability of heads. Each coin will be fair
p = 0.5

# Flip two coins several times.
nflips = np.random.binomial(n, p, k)

# Count the number of times each value occurs.
uniques, counts = np.unique(nflips, return_counts=True)

# Create a blank plot.
fig, ax = plt.subplots(figsize=(12, 5))

# Create a bar chart.
# what proportion of the time each number of heads occurs
# its the proportion because we divide by k, rather than the number of times
ax.bar(uniques, counts / k)

# Create an x range.
# we saw before that values below 25 and above 75 were very rare
# use linspace to create an array of values between 25 and 75
# use 1001 points to get a smooth curve
x = np.linspace(25.0, 75.0, 1001)

# Get y values.
# p is the probability of heads
# 1 minus p is the probability of tails
# in this example, they are equal as the coin is fair at 0.5
# using this function using reasonable values will result in the normal distribution curve
# refers cne to bool above is where Iam directs usd to this function
y = f(x, mu=n*p, sigma=np.sqrt(n * p * (1 - p)))

# Plot the function.
ax.plot(x, y, 'r-')

# Set x ticks.
ax.set_xticks(uniques)

# Add title.
ax.set_title(f'Flipping {n} coins {k} times')

# Add axis label.
ax.set_xlabel('No heads')
ax.set_ylabel('Proportion');
```

### Problem 2

Generate 100,000 samples of size 10 from the standard normal distribution.
For each sample, compute the standard deviation with `ddof=1` (sample SD) and with `ddof=0` (population SD).
Plot histograms of both sets of values on the same axes with transparency.
Describe the differences you see.
Explain how you expect these differences to change if the sample size is increased.


using density

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_samples = 100000
sample_size = 10

# Generate random samples
samples = np.random.randn(n_samples, sample_size)

# Compute standard deviations
sd_ddof0 = samples.std(axis=1, ddof=0)  # population SD
sd_ddof1 = samples.std(axis=1, ddof=1)  # sample SD

# Plot histograms
plt.hist(sd_ddof0, bins=50, alpha=0.5, label='ddof=0 (population SD)', density=True)
plt.hist(sd_ddof1, bins=50, alpha=0.5, label='ddof=1 (sample SD)', density=True)
plt.xlabel('Standard deviation')
plt.ylabel('Density')
plt.title('Distribution of sample vs population standard deviations (n=10)')
plt.legend()
plt.show()
```

Using frequency

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_samples = 100000  # number of repetitions
sample_size = 10     # number of data points in each sample

# Generate 100,000 samples of size 10 from N(0, 1)
samples = np.random.randn(n_samples, sample_size)

# Compute standard deviations
sd_sample = np.std(samples, axis=1, ddof=1)  # sample SD
sd_population = np.std(samples, axis=1, ddof=0)  # population SD

# Plot histograms
plt.hist(sd_sample, bins=50, alpha=0.5, label='Sample SD (ddof=1)')
plt.hist(sd_population, bins=50, alpha=0.5, label='Population SD (ddof=0)')
plt.xlabel('Standard Deviation')
plt.ylabel('Frequency')
plt.legend()
plt.title('Sampling Distribution of Standard Deviations')
plt.show()

```

using counts

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_samples = 100_000
sample_size = 10

# Generate 100,000 samples of size 10 from N(0, 1)
samples = np.random.randn(n_samples, sample_size)

# Compute standard deviations
sd_ddof0 = np.std(samples, axis=1, ddof=0)  # population SD
sd_ddof1 = np.std(samples, axis=1, ddof=1)  # sample SD

# Use the same bins for both histograms so they align
vmin = min(sd_ddof0.min(), sd_ddof1.min())
vmax = max(sd_ddof0.max(), sd_ddof1.max())
bins = np.linspace(vmin, vmax, 50)

# Plot raw counts (no density normalization)
plt.hist(sd_ddof0, bins=bins, linewidth=2,label='Population SD (ddof=0)', color='salmon')
plt.hist(sd_ddof1, bins=bins, linewidth=2,label='Sample SD (ddof=1)', color='skyblue')

plt.xlabel('Standard Deviation')
plt.ylabel('Frequency (Count of samples)')
plt.title(f'Sampling distribution of SDs (n={sample_size}, {n_samples} repetitions)')
plt.legend()
plt.show()
```
