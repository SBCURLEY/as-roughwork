```python
# Flip a single coin several times.
flips = np.random.binomial(1, 0.5, 1000)  # 1000 flips of a fair coin

# Show.
flips
```

```python
# Count heads.  heads is a 1 and tails is a 0
heads = flips.sum()   # takes teh flips array and sums it up    .sum() is in numpy array    

# Show.
heads
```

#### 1000 Coins
Instead of flipping one coin 1000 times, flip 1000 coins once

You are simulating the total number of heads from 1000 coin flips. The value stored in flips is already an integer that represents the number of heads you got in those 1000 flips (for example, 500 heads). There's no need to sum anything because flips is already the sum.
So, in this case, flips is already the result you want — you don't need to sum it again. You can simply use flips directly as the number of heads.

```python
np.random.binomial(1000, 0.5) 
```

```python
flips = np.random.binomial(1000, 0.5)  
# Show.
print(f"In this simulation, out of 1000 flips, {flips} were heads.")
```

Clarification:

To recap:

np.random.binomial(1000, 0.5) gives you a single integer, which is the number of heads (successes) from 1000 trials.

If you wanted to simulate each individual flip (0 or 1 for each coin flip), you’d use np.random.binomial(1, 0.5, 1000).

## Probability Density Function

$ f(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $

https://en.wikipedia.org/wiki/Normal_distribution

```python
def f(x, mu=0.0, sigma=1.0):
  # The normalisation constant.
  norm = 1.0 / np.sqrt(2.0 * np.pi * sigma**2)
  
  # f(x)
  return norm * np.exp(-(x - mu)**2 / (2.0 * sigma**2))
```

```python
# Test the function.
f(0.0)
```

```python
# Create an x range.
x = np.linspace(-5, 5, 1001)

# Get y values.
y = f(x)

# Show.
x, y
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Plot the function.
ax.plot(x, y);
```

Reference: [6.4: Normal Approximation to the Binomial Distribution](https://stats.libretexts.org/Courses/Las_Positas_College/Math_40%3A_Statistics_and_Probability/06%3A_Continuous_Random_Variables_and_the_Normal_Distribution/6.04%3A_Normal_Approximation_to_the_Binomial_Distribution)

```python
# Number of coins to flip.
n = 100

# Number of times to do it.
k = 10000 * n

# Probability of heads.
p = 0.5

# Flip two coins several times.
nflips = np.random.binomial(n, p, k)

# Count the number of times each value occurs.
uniques, counts = np.unique(nflips, return_counts=True)

# Create a blank plot.
fig, ax = plt.subplots(figsize=(12, 5))

# Create a bar chart.
ax.bar(uniques, counts / k)

# Create an x range.
x = np.linspace(25, 75, 1001)

# Get y values.
y = f(x, mu=n*p, sigma=np.sqrt(n * p * (1 - p)))

# Plot the function.
ax.plot(x, y, 'r-')

# Set x ticks.
ax.set_xticks(uniques)

# Add title.
ax.set_title(f'Flipping {n} coins {k} times')

# Add axis label.
ax.set_xlabel('No heads')
ax.set_ylabel('Proportion');
```

## Integration

https://en.wikipedia.org/wiki/97.5th_percentile_point

```python
# Adapted from: https://stackoverflow.com/a/10138308

# Create a blank plot.
fig, ax = plt.subplots()

# X values.
x = np.linspace(-4.0, 4.0, 1000)

# Plot the normal distribution.
plt.plot(x, stats.norm.pdf(x))

# Shade in some areas.
plt.fill_between(x, stats.norm.pdf(x), where=(x < -1.96) | (x > 1.96), color='red', alpha=0.7);
```
